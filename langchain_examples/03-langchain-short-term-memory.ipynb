# Langchain Short Term Memory

```python
from dotenv import load_dotenv
import os

load_dotenv()
os.environ["OPEN_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(model="gpt-5.4-mini")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from langchain.messages import HumanMessage

agent = create_agent(model=llm, tools=[], system_prompt="You are a helpful assistant that can answer questions about me.", checkpointer=InMemorySaver())

question = HumanMessage(content="Hello My name is Shubham and my favorite color is blue")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke({"messages": [question]}, config)

print(response)
```

```python
question = HumanMessage(content="What is my favorite color?")

response = agent.invoke({"messages": [question]}, config)

print(response["messages"][-1].content)
```
